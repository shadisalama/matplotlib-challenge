```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_drug_data_to_load_df = pd.read_csv('data/mouse_drug_data.csv')

#mouse_drug_data_to_load_df

clinical_trial_data_to_load_df = pd.read_csv('data/clinicaltrial_data.csv')
#clinical_trial_data_to_load_df

# Combine the data into a single dataset

merge_table = pd.merge(mouse_drug_data_to_load_df, clinical_trial_data_to_load_df, on="Mouse ID")


# Display the data table for preview

merge_table
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

tumor_volume_mean = merge_table.groupby(["Drug","Timepoint"]).mean()["Tumor Volume (mm3)"]
#tumor_volume_mean

# Convert to DataFrame

tumor_volume_df = pd.DataFrame ({"Tumor Volume (mm3)": tumor_volume_mean})

# Preview DataFrame
tumor_volume_df = tumor_volume_df.reset_index()
tumor_volume_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

tumor_volume_sem = merge_table.groupby(["Drug","Timepoint"]).sem()["Tumor Volume (mm3)"]
tumor_volume_sem

# Convert to DataFrame
tumor_volume_sem_df = pd.DataFrame ({"Tumor Volume (mm3)": tumor_volume_sem})

# Preview DataFrame

tumor_volume_sem_df = tumor_volume_sem_df.reset_index()
tumor_volume_sem_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_mean = tumor_volume_df.pivot(index='Timepoint',columns='Drug')['Tumor Volume (mm3)']

# Preview that Reformatting worked
pivot_mean.head()
```

```python
pivot_sem = tumor_volume_sem_df.pivot(index='Timepoint',columns='Drug')['Tumor Volume (mm3)']
pivot_sem.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(pivot_mean.index,pivot_mean['Capomulin'],yerr=pivot_sem['Capomulin'],marker='o',color='r',label='Capomulin')
plt.errorbar(pivot_mean.index,pivot_mean['Placebo'],yerr=pivot_sem['Placebo'],marker='^',color='b',label='Placebo')
plt.errorbar(pivot_mean.index,pivot_mean['Infubinol'],yerr=pivot_sem['Infubinol'],marker='s',color='g',label='Infubinol')
plt.errorbar(pivot_mean.index,pivot_mean['Ketapril'],yerr=pivot_sem['Ketapril'],marker='d',color='k',label='Ketapril')

# Save the Figure
plt.title('Tumor Response to Treatment')
plt.xlabel('Time(Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Minor Data Munging to Re-Format the Data Frames

pivot_metastatic_response_mean = metastatic_response_df.pivot(index='Timepoint',columns='Drug')['Metastatic Sites']

# Preview that Reformatting worked
pivot_metastatic_response_mean.head()
```

```python
# Generate the Plot (with Error Bars)

plt.errorbar(pivot_metastatic_response_mean.index,pivot_metastatic_response_mean['Capomulin'],yerr=pivot_sem['Capomulin'],marker='o',color='r',label='Capomulin')
plt.errorbar(pivot_metastatic_response_mean.index,pivot_metastatic_response_mean['Placebo'],yerr=pivot_sem['Placebo'],marker='d',color='k',label='Placebo')
plt.errorbar(pivot_metastatic_response_mean.index,pivot_metastatic_response_mean['Infubinol'],yerr=pivot_sem['Infubinol'],marker='^',color='b',label='Infubinol')
plt.errorbar(pivot_metastatic_response_mean.index,pivot_metastatic_response_mean['Ketapril'],yerr=pivot_sem['Ketapril'],marker='s',color='g',label='Ketapril')


# Save the Figure

# Show the Figure

plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treatment Duration(Days)')
plt.ylabel('Met. Sites')
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count = merge_table.groupby(["Drug","Timepoint"]).count()["Mouse ID"]
mouse_count
# Convert to DataFrame
mouse_count_df = pd.DataFrame({"Mouse Count":mouse_count})

# Preview DataFrame
mouse_count_df = mouse_count_df.reset_index()
mouse_count_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count_pivot = mouse_count_df.pivot(index='Timepoint',columns='Drug')['Mouse Count']
mouse_count_pivot
# Preview the Data Frame
```

```python
# Generate the Plot (Accounting for percentages)

plt.plot(mouse_count_pivot.index,mouse_count_pivot['Capomulin'],marker='o',color='r',label='Capomulin')
plt.plot(mouse_count_pivot.index,mouse_count_pivot['Placebo'],marker='d',color='k',label='Placebo')
plt.plot(mouse_count_pivot.index,mouse_count_pivot['Infubinol'],marker='^',color='b',label='Infubinol')
plt.plot(mouse_count_pivot.index,mouse_count_pivot['Ketapril'],marker='s',color='g',label='Ketapril')


# Save the Figure

# Show the Figure
plt.title('Survival During Treatment')
plt.xlabel('Time(Days)')
plt.ylabel('Survival Rate (%)')
plt.grid()
plt.legend(loc='lower left')
plt.show()
```
